```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris = iris.loc[iris.species == "setosa"][["sepal_length", "sepal_width"]]
iris.head(2)
```

```python
fig = plt.figure(figsize = (6, 6))
ax = plt.axes(aspect = "equal", xlim = (0, 7), ylim = (0, 7))
sns.scatterplot(x = "sepal_length", y = "sepal_width", data = iris, ax = ax)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 1, random_state = 0)
```

```python
iris_transformed = pca.fit_transform(iris)
```

```python
iris_transformed[:5, :]
```

```python
sns.scatterplot(x = iris_transformed[:, 0], y = [0] * len(iris_transformed), ax = ax, color = "red")
```

```python
iris_o_transformed = pca.inverse_transform(iris_transformed)
```

```python
fig = plt.figure(figsize = (6, 6))
ax = plt.axes(aspect = "equal", xlim = (0, 7), ylim = (0, 7))
sns.scatterplot(x = "sepal_length", y = "sepal_width", data = iris, ax = ax)
sns.scatterplot(x = iris_o_transformed[:, 0], y = iris_o_transformed[:, 1], ax = ax, color = "red")
```
